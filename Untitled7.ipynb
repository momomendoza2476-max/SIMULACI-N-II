```python
import numpy as np
import matplotlib.pyplot as plt

# Tu función f(x1, x2)
def f(x1, x2):
    return np.exp(-(x1**2 * x2**2 + x1**2 + x2**2 - 8*x1 - 8*x2) / 2)

# Probabilidad de aceptación
def a(x, y):
    fx = f(x[0], x[1])
    fy = f(y[0], y[1])
    return min(1, fy/fx)

# Propuesta normal alrededor del punto actual
def norm(mu, s=1):
    return np.random.normal(mu, s, size=2)

# Algoritmo de Metropolis-Hastings
def MH(x0, n):
    X = [x0]
    for i in range(n):
        y = norm(X[-1])
        u = np.random.uniform(0, 1)
        if u < a(X[-1], y):
            X.append(y)
        else:
            X.append(X[-1])
    return np.array(X)

# Ejecutamos
muestras = MH(np.array([0,0]), 100000)

# Gráfica de dispersión
plt.figure(figsize=(7,6))
plt.hist2d(muestras[:,0], muestras[:,1], bins=100, cmap='viridis', density=True)
plt.colorbar(label='Densidad aproximada')
plt.title('Muestras generadas por Metropolis-Hastings')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Función objetivo ---
def f(x1, x2):
    return np.exp(-0.5 * (x1**2 * x2**2 + x1**2 + x2**2 - 8*x1 - 8*x2))

# --- Parámetros del muestreo ---
n_samples = 50000     # número total de muestras
burn_in = 5000         # descartar las primeras muestras
sigma = 0.5            # desviación estándar de la propuesta

# --- Inicialización ---
x1_chain = np.zeros(n_samples)
x2_chain = np.zeros(n_samples)
x1_chain[0], x2_chain[0] = 0.0, 0.0  # punto inicial

# --- Algoritmo de Metropolis-Hastings ---
for i in range(1, n_samples):
    x1_prop = np.random.normal(x1_chain[i-1], sigma)
    x2_prop = np.random.normal(x2_chain[i-1], sigma)
    r = f(x1_prop, x2_prop) / f(x1_chain[i-1], x2_chain[i-1])
    if np.random.rand() < min(1, r):
        x1_chain[i], x2_chain[i] = x1_prop, x2_prop
    else:
        x1_chain[i], x2_chain[i] = x1_chain[i-1], x2_chain[i-1]

print("Muestreo completado.")

# --- Descartar fase de burn-in ---
x1_samples = x1_chain[burn_in:]
x2_samples = x2_chain[burn_in:]

# --- Crear malla para f(x1, x2) ---
x = np.linspace(-2, 10, 150)
y = np.linspace(-2, 10, 150)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# --- Estimar densidad de los puntos muestreados ---
hist, xedges, yedges = np.histogram2d(x1_samples, x2_samples, bins=60, range=[[-2,10],[-2,10]], density=True)
xpos, ypos = np.meshgrid((xedges[:-1] + xedges[1:])/2,
                         (yedges[:-1] + yedges[1:])/2)
Z_est = hist.T

# --- Activar modo interactivo ---
plt.ion()

# --- Gráficas 3D ---
fig = plt.figure(figsize=(14,6))

ax1 = fig.add_subplot(1,2,1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, alpha=0.8)
ax1.set_title('Función teórica $f(x_1, x_2)$')
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_zlabel('$f(x_1, x_2)$')

ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.plot_surface(xpos, ypos, Z_est, cmap='plasma', linewidth=0, alpha=0.8)
ax2.set_title('Densidad estimada (Metrópolis–Hastings)')
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')
ax2.set_zlabel('Densidad estimada')

plt.tight_layout()
plt.show()
```
