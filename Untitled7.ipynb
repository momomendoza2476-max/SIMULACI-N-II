```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # importa herramientas 3D

# Activa el modo interactivo (solo en notebooks o entornos gráficos)
plt.ion()

# Constante
C = 1 / 20216.335877

# Función
def f(x1, x2):
    return C * np.exp(-(x1**2 * x2**2 + x1**2 + x2**2 - 8*x1 - 8*x2) / 2)

# Rango de valores
x = np.linspace(-2, 10, 150)
y = np.linspace(-2, 10, 150)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Gráfica 3D interactiva
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')

# Superficie
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

# Etiquetas
ax.set_title('Superficie 3D interactiva de f(x₁, x₂)')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.set_zlabel('f(x₁, x₂)')

# Barra de color
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Muestra la figura interactiva
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tu función f(x1, x2)
def f(x1, x2):
    return np.exp(-(x1**2 * x2**2 + x1**2 + x2**2 - 8*x1 - 8*x2) / 2)

# Probabilidad de aceptación
def a(x, y):
    fx = f(x[0], x[1])
    fy = f(y[0], y[1])
    return min(1, fy/fx)

# Propuesta normal alrededor del punto actual
def norm(mu, s=1):
    return np.random.normal(mu, s, size=2)

# Algoritmo de Metropolis-Hastings
def MH(x0, n):
    X = [x0]
    for i in range(n):
        y = norm(X[-1])
        u = np.random.uniform(0, 1)
        if u < a(X[-1], y):
            X.append(y)
        else:
            X.append(X[-1])
    return np.array(X)

# Ejecutamos
muestras = MH(np.array([0,0]), 100000)

# Gráfica de dispersión
plt.figure(figsize=(7,6))
plt.hist2d(muestras[:,0], muestras[:,1], bins=100, cmap='viridis', density=True)
plt.colorbar(label='Densidad aproximada')
plt.title('Muestras generadas por Metropolis-Hastings')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
