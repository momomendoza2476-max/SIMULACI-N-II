```python
from random import expovariate
from statistics import mean

# Parámetros
lamda = 0.8       # Tasa de llegadas
mu = 1.0          # Tasa de servicio
Num_Pkts = 100000 # Número de paquetes a simular

# Inicialización
count = 0
t = 0.0
N = 0   # Número de clientes en el sistema

tll = expovariate(lamda)  # tiempo de siguiente llegada
ts = float('inf')         # tiempo de siguiente salida
t_evento_anterior = 0.0

# Variables de salida
tll_Data = []   # tiempos de llegada
ts_Data = []    # tiempos de salida
r_Data = []     # retrasos individuales

# Diccionario para acumular tiempo en cada estado N
Data = {}

# -------------------------------
#           SIMULACIÓN
# -------------------------------

while count < Num_Pkts:

    # Evento de llegada
    if tll < ts:
        t = tll

        # Longitud del intervalo desde el último evento
        delta = t - t_evento_anterior

        # Acumular tiempo en estado N
        if N in Data:
            Data[N] += delta
        else:
            Data[N] = delta

        t_evento_anterior = t

        # Registrar llegada
        tll_Data.append(t)
        N += 1

        # Nueva llegada
        tll = t + expovariate(lamda)

        # Si el sistema estaba vacío, programamos una salida
        if N == 1:
            ts = t + expovariate(mu)

    # Evento de salida
    else:
        t = ts

        # Longitud del intervalo desde el último evento
        delta = t - t_evento_anterior

        if N in Data:
            Data[N] += delta
        else:
            Data[N] = delta

        t_evento_anterior = t

        # Registrar salida
        ts_Data.append(t)
        N -= 1
        count += 1

        # Si el sistema queda vacío
        if N == 0:
            ts = float('inf')
        else:
            ts = t + expovariate(mu)

# -------------------------------
#     CÁLCULO DE RETRASOS
# -------------------------------

# Asegurarse de que tll_Data y ts_Data tengan el mismo largo (deberían)
min_len = min(len(tll_Data), len(ts_Data))
for i in range(min_len):
    d = ts_Data[i] - tll_Data[i]
    r_Data.append(d)

print("Retraso promedio = ", round(mean(r_Data), 4))

# -------------------------------
#  ESTIMACIÓN DE P(N = k)
# -------------------------------

# Normalización para obtener probabilidades
total_time = sum(Data.values())

for key, value in list(Data.items()):
    Data[key] = value / total_time

print("Suma de probabilidades =", sum(Data.values()))

# Valor esperado E[N]
media = 0.0
for key, p in Data.items():
    media += key * p

print("E[N] = ", media)

# Mostrar distribución (opcional)
print("\nDistribución estimada P(N=k):")
for key in sorted(Data.keys()):
    print(key, round(Data[key], 4))
```
