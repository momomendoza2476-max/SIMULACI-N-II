```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
def g(x):
  return np.sqrt(np.arctan(x))
```

```python
X=np.linspace(0,1,100)
plt.plot(X,g(X))
```

```python
def crudo(N):
  G=[]
  for i in range (N):
    u=rd.random()
    G.append(g(u))
  return np.mean(G), np.var(G), np.std(G), G
```

```python
e=crudo(100)
```

```python
e[0]
```

```python
print("El valor estimado es: ", e[0])
print("La varianza es: ", e[1])
print("La desviación estándar es: ", e[2])
```

```python
L=[]
for i in range (1000):
  z=crudo(i)
  L.append(z[0]) #recordar que la función crudo devuelve un arreglo, necesitamos la media.
```

```python
m=np.mean(L)
sigma=np.std(L)
```

```python
plt.plot(L)
plt.title("Estimación de $I$ vs N")
plt.xlabel("Tamaño de la muetra")
plt.ylabel("Estimación de $I$")
plt.axvline(500, color="red")
plt.axhline(m, color="red")
plt.axhline(m+sigma, color="green")
plt.axhline(m-sigma, color="green")
plt.annotate("Región estable", xy=(800,0,75), color="red")
plt.annotate("Valor de la integral", xy=(800,0.64))
```
