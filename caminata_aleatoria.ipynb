CAMINATA ALEATORIA 1D

```python
import random
import matplotlib.pyplot as plt

# Número de pasos
n_pasos = 1000

# Posiciones
posicion = [0]

# Generar caminata 1D
for _ in range(n_pasos):
    paso = random.choice([-1, 1])  # Subir o bajar
    nueva_pos = posicion[-1] + paso
    posicion.append(nueva_pos)

# Gráfica
plt.figure(figsize=(10, 5))
plt.plot(posicion, marker='o', markersize=2, linestyle='--', linewidth=1)
plt.title("Caminata Aleatoria en 1D (partiendo desde 0)")
plt.xlabel("Número de pasos")
plt.ylabel("Posición")
plt.axhline(0, color='red', linestyle='--')
plt.grid(True, linestyle=':')
plt.show()
```

Caminata aleatoria 2D

```python
import random
import matplotlib.pyplot as plt

# Número de pasos
n_pasos = 1000

# Posiciones iniciales
x = [0]
y = [0]

# Generar caminata aleatoria en 2D
for _ in range(n_pasos):
    dx = random.choice([-1, 1])
    dy = random.choice([-1, 1])
    x.append(x[-1] + dx)
    y.append(y[-1] + dy)

# Graficar caminata 2D
plt.figure(figsize=(8, 8))
plt.plot(x, y, marker='o', markersize=2, linestyle='-')
plt.title('Caminata Aleatoria en 2D')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axis('equal')  # Para mantener la proporción
plt.show()
```

CAMINATA ALEATORIA 3D



```python
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import necesario para gráficas 3D

# Número de pasos
n_pasos = 1000

# Posiciones iniciales
x = [0]
y = [0]
z = [0]

# Generar caminata en 3D
for i in range(1, n_pasos + 1):
    paso_x = random.choice([-1, 1])
    paso_y = random.choice([-1, 1])
    paso_z = random.choice([-1, 1])

    x.append(x[-1] + paso_x)
    y.append(y[-1] + paso_y)
    z.append(z[-1] + paso_z)

# Gráfica 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, marker='o', linestyle='-', linewidth=1, markersize=2)

ax.set_title('Caminata Aleatoria en 3D')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.show()
```
