```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw

def modelo_sir_simplificado():
    print("MODELO TEÓRICO EPIDÉMICO (versión Lambert W)")

    # Parámetros del modelo
    a = 2
    N = 1000
    X0 = 1

    # Definir la función X(R)
    def X(R):
        return N - R - (N - X0) * np.exp(-a * R / N)

    # Calcular Rinf usando la función W de Lambert
    Rinf = N * (1 + lambertw(-a * np.exp(-a) * (1 - X0 / N)) / a)
    print(f'El valor de R donde X(R) = 0 es R ≈ {Rinf:.4f}')

    # Crear la gráfica
    plt.figure(figsize=(10, 6))
    Rvals = np.arange(0, int(Rinf) + 1)
    plt.plot(Rvals, X(Rvals), 'r-', linewidth=1.5)
    plt.plot([0, N], [0, 0], 'k--', linewidth=1)
    plt.grid(True)
    plt.xlabel('R')
    plt.ylabel('X(R)')
    plt.title('Modelo teórico SIR simplificado con parámetro a')

    # Marcar la solución en la gráfica
    plt.plot(Rinf, X(Rinf), 'bo', markerfacecolor='b', markersize=8)
    plt.text(Rinf, 0, f' R_∞ ≈ {Rinf:.2f}', verticalalignment='bottom')

    # Calcular el máximo de X
    RXmax = N * np.log(a * (1 - X0 / N)) / a
    Xmax = X(RXmax)
    print(f'RXmax = {RXmax:.4f}')
    print(f'Xmax = {Xmax:.4f}')

    plt.tight_layout()
    plt.show()

    return Rinf, RXmax, Xmax

def simulacion_estocastica():
    """
    Simulación estocástica del modelo SIR
    """
    # Parámetros
    a = 2
    N = 1000
    x = 1 # X0
    r = 0

    print("\nSimulación estocástica:")
    pasos = 0
    historial_x = [x]
    historial_r = [r]

    while x > 0 and pasos < 10000: # Límite de pasos por seguridad
        # Tiempos de los eventos
        if x > 0:
            dtSI = -np.log(np.random.random()) / (a * (1 - (r + x) / N)) / x
        else:
            dtSI = float('inf')

        if x > 0:
            dtIR = -np.log(np.random.random()) / x
        else:
            dtIR = float('inf')

        # Determinar qué evento ocurre primero
        if dtIR < dtSI: # Alguien se cura
            r += 1
            x -= 1
        else: # Alguien se infecta
            x += 1

        historial_x.append(x)
        historial_r.append(r)
        pasos += 1

    print(f"Simulación terminada después de {pasos} pasos")
    print(f"R final: {r}, X final: {x}")

    # Graficar la simulación estocástica
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(historial_x, 'b-', label='X (Susceptibles)', linewidth=1)
    plt.plot(historial_r, 'r-', label='R (Recuperados)', linewidth=1)
    plt.xlabel('Pasos de tiempo')
    plt.ylabel('Población')
    plt.title('Simulación estocástica SIR')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(historial_r, historial_x, 'g-', linewidth=1)
    plt.xlabel('R (Recuperados)')
    plt.ylabel('X (Susceptibles)')
    plt.title('Trayectoria en el plano X-R')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Ejecutar el modelo determinista
if __name__ == "__main__":
    Rinf, RXmax, Xmax = modelo_sir_simplificado()

    # Opcional: ejecutar simulación estocástica
    # simulacion_estocastica()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw

def modelo_sir_simplificado():
    print("MODELO TEÓRICO EPIDÉMICO (versión Lambert W)")

    # Parámetros del modelo
    a = 2
    N = 1000
    X0 = 1

    # Definir la función X(R)
    def X(R):
        return N - R - (N - X0) * np.exp(-a * R / N)

    # Calcular Rinf usando la función W de Lambert (tomando parte real)
    Rinf_complex = N * (1 + lambertw(-a * np.exp(-a) * (1 - X0 / N)) / a)
    Rinf = np.real(Rinf_complex) # Tomar solo la parte real
    print(f'El valor de R donde X(R) = 0 es R ≈ {Rinf:.4f}')

    # Crear la gráfica
    plt.figure(figsize=(10, 6))
    Rvals = np.arange(0, int(Rinf) + 1)
    plt.plot(Rvals, X(Rvals), 'r-', linewidth=1.5)
    plt.plot([0, N], [0, 0], 'k--', linewidth=1)
    plt.grid(True)
    plt.xlabel('R')
    plt.ylabel('X(R)')
    plt.title('Modelo teórico SIR simplificado con parámetro a')

    # Marcar la solución en la gráfica (punto azul)
    plt.plot(Rinf, X(Rinf), 'bo', markerfacecolor='b', markersize=8)
    plt.text(Rinf, 0, f' R_∞ ≈ {Rinf:.2f}', verticalalignment='bottom')

    # Calcular el máximo de X
    RXmax = N * np.log(a * (1 - X0 / N)) / a
    Xmax = X(RXmax)
    print(f'RXmax = {RXmax:.4f}')
    print(f'Xmax = {Xmax:.4f}')

    # También marcar el punto máximo en la gráfica
    plt.plot(RXmax, Xmax, 'go', markerfacecolor='g', markersize=8)
    plt.text(RXmax, Xmax, f' Máximo\n R={RXmax:.1f}\n X={Xmax:.1f}',
             verticalalignment='bottom')

    plt.tight_layout()
    plt.show()

    return Rinf, RXmax, Xmax

# Instalar bibliotecas si no las tienes
def instalar_bibliotecas():
    """
    Para instalar las bibliotecas necesarias, ejecuta en la terminal:

    pip install numpy matplotlib scipy
    """
    print("Para instalar las bibliotecas, ejecuta en terminal:")
    print("pip install numpy matplotlib scipy")

# Ejecutar el modelo
if __name__ == "__main__":
    # Si necesitas instalar las bibliotecas primero:
    # instalar_bibliotecas()

    Rinf, RXmax, Xmax = modelo_sir_simplificado()
```
