```python
import random
import numpy as np
from collections import Counter

lmbda = 2.0
mu = 3.0
tiempo_total = 5000

t = 0
num_clientes = 0
registro_n = []

#medir tiempos
tiempo_llegada = []       #llegada de cada cliente
tiempo_inicio_servicio = []
tiempo_salida = []

t_llegada = random.expovariate(lmbda)
t_salida = float('inf')

while t < tiempo_total:

    if t_llegada < t_salida:

        t = t_llegada
        num_clientes += 1

        registro_n.append(num_clientes)
        tiempo_llegada.append(t)


        t_llegada = t + random.expovariate(lmbda)

        #si el cliente entra directo al servicio
        if num_clientes == 1:
            tiempo_inicio_servicio.append(t)  #entra directo
            t_salida = t + random.expovariate(mu)

    else:

        t = t_salida
        num_clientes -= 1

        registro_n.append(num_clientes)
        tiempo_salida.append(t)

        #si quedan clientes esperando
        if num_clientes > 0:
            tiempo_inicio_servicio.append(t)  #siguiente empieza servicio
            t_salida = t + random.expovariate(mu)
        else:
            t_salida = float('inf')


# Cálculo de probabilidades P(N=n)
from collections import Counter
conteo = Counter(registro_n)
total = len(registro_n)

print("Probabilidades estimadas:")
for n in sorted(conteo.keys()):
    print(f"P(N={n}) ≈ {conteo[n] / total:.4f}")


#cálculo de tiempos W y Wq

#listas de igual tamaño
n = min(len(tiempo_llegada), len(tiempo_inicio_servicio), len(tiempo_salida))

llegadas = np.array(tiempo_llegada[:n])
inicios = np.array(tiempo_inicio_servicio[:n])
salidas = np.array(tiempo_salida[:n])

#W = tiempo total que pasa un cliente en el sistema
W = np.mean(salidas - llegadas)

#Wq = tiempo que pasa en la fila
Wq = np.mean(inicios - llegadas)

print("\n--- Tiempos promedio ---")
print(f"Tiempo promedio en el sistema  W ≈ {W:.4f}")
print(f"Tiempo promedio en la fila     Wq ≈ {Wq:.4f}")
```
